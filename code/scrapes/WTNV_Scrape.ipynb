# Find Episode Links and Scrape Them

```python
#get all links
divparent = main_html.find_all('li')
```

```python
#find hrefs with links to episodes
#pull all from html li that match regex
result = re.findall(r"<a\shref=\"\/[\w\-\:]*", str(divparent))
#drop last 5 entries (rss,random,archive,ask, and mobile)
result = result[:-5]
#get rid of beginning stuff
result = [r[10:] for r in result]
```

```python
#additional urls
urls = ['https://cecilspeaks.tumblr.com/post/182482146001/141-save-dark-owl-records', 
        'https://cecilspeaks.tumblr.com/post/182853449756/142-ufo-sighting-reports', 
        'https://cecilspeaks.tumblr.com/post/183149396746/143-pioneer-days']
k = ['ep141','ep142','ep143']
```

```python
#create dictionary for html
wtnv_html_dict = dict()

#in loop, tack link onto main url
for r in result:
    raw_string = main_url+r
    key = r
    print(key, raw_string, time.localtime())
    #pull in raw html
    time.sleep(1)
    tz = simple_get(raw_string)
    wtnv_html_dict[key] = tz
```

```python
for i in range(len(urls)):
    time.sleep(1)
    raw_string = urls[i]
    key = k[i]
    print(key, raw_string, time.localtime())
    #pull in raw html
    tz = simple_get(raw_string)
    wtnv_html_dict[key] = tz
```

# Clean Data and Make DataFrame

```python
cleaned = pd.DataFrame()
eptypes = []
epnums = []
epnames = []
texts = []
```

```python
for entry in wtnv_html_dict:
    #BeautifulSoup to html
    html = BeautifulSoup(wtnv_html_dict[entry], 'html.parser')
    title = str(html.find("meta",  property="og:title"))
    markup = html.find("meta",  property="og:description").get('content').encode("utf-8")
    content = markup.decode("utf-8", "ignore")
    content = content.replace('\n','')
    
    if len(title.split('\"')) == 1:
        title = 'The Librarian (Preview) featuring Horoscopes'
    else:
        title = title.split('\"')[1]
    
    if 'Bonus' in title:
            ep_type = 'bonus'

            diglist = re.findall(r"[0-9]+",title)
            tlist = title.split('-')
            
            if len(diglist) > 0:
                ep_num = int(diglist[0])
            else:
                ep_num = -1
                
            if len(tlist) > 1:
                ep_name = tlist[1][1:]
            else:
                ep_name = title
                
    elif ('Live ' in title) or ('The Investigators' == title) or ('Librarian' in title):
        ep_type = 'live'
        ep_num = -1
        if 'Librarian' in title:
            ep_name = 'The Librarian'
        elif '2017' in title:
            ep_name = 'ALL HAIL'
        else:
            ep_name = title
            
    elif ('Condos' in title) or ('The Debate' == title) or ('Crossover' in title):
        ep_type = 'live'
        ep_num = -1
        if 'Condos' in title:
            ep_name = 'Condos'
        else:
            ep_name = title
            
    else:
        ep_type = 'regular'
        diglist = re.findall(r"[0-9]+",title)
        ep_num = int(diglist[0])
        tlist = title.split('-')
            
        if len(tlist) > 1:
            ep_name = tlist[1][1:]
        else:
            ep_name = title.split(':')[1]
                
        
    eptypes.append(ep_type)
    epnums.append(ep_num)
    epnames.append(ep_name)
    
    if (len(content) == 0):
        plist = html.find_all("p")
        for ind in range(76):
            p = plist[ind]
            s = str(p).replace('<p>',' ').replace('</p>',' ').replace('<i>',' ').replace('</i>',' ')
            s = s.replace('<b>',' ').replace('</b>',' ').replace('<br/>',' ')
            if ind != 63:
                content = content + ' ' + s
                
    
    texts.append(content)
```

```python
d = {'episode_type': eptypes, 'episode_num': epnums, 'episode_name': epnames, 'text':texts}
df = pd.DataFrame(data=d)
```

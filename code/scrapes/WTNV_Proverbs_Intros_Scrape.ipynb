# Scrape

```python
#scrape page with all intros and proverbs
url = 'https://nightvale.fandom.com/wiki/Introductions_and_Proverbs'
tz = simple_get(url)
time.sleep(1)

#parse html
html = BeautifulSoup(tz, 'html.parser')
```

# Format

```python
#all info is in table, pull out table
table = html.find("table")
eps = []
intros = []
proverbs = []

#for each row
for row in table.findAll("tr")[1:]:
    #get all the cells
    cells = row.findAll("td")
    
    #extract episode name
    ep = str(cells[0].renderContents()).strip().replace('\\n','').replace('<p>','').replace('</p>','')[2:-1]
    eps.append(ep)
    
    #extract intro
    intro = str(cells[1].renderContents()).strip().replace('\\n','').replace('<p>','').replace('</p>','')[2:-1]
    intros.append(intro)
    
    #extract proverb
    proverb = str(cells[2].renderContents().decode("utf-8", "ignore")).replace('\n',' ').replace('<br/>','')
    proverb = proverb.replace('<p>','').replace('</p>','').replace('<ol>','').replace('<li>','')
    proverb = proverb.replace('</li>','').replace('</ol>','').replace('<i>','').replace('</i>','').replace('< /b>','')
    proverb = proverb.replace('<b>','').replace('``','')
    proverbs.append(proverb)
```

```python
#make dataframe
d = {'Episode': eps, 'Intro': intros, 'Proverb': proverbs}
df = pd.DataFrame(data=d)
```
